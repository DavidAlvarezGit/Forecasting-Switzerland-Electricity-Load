```python
import pandas as pd
from pathlib import Path


def xml_to_parquet_easy(xml_file):
    xml_file = Path(xml_file)

    df = pd.read_xml(xml_file, xpath=".//Point")

    out_file = xml_file.with_suffix(".parquet")

    df.to_parquet(out_file, index=False)

    print(f"Saved: {out_file}")

    return df
```
