```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_parquet('../../data/raw/entsoe/swiss_load_2020.parquet')

df.set_index('timestamp_utc', inplace=True)

df[df.isna().any(axis=1)]

df.index = pd.to_datetime(df.index)

df.info()


```

```python
def negative_load(df):
    assert (df['load_mw'] >= 0).all()
```

```python
def check_no_missing_hours(df):
    expected = pd.date_range(df.index.min(), df.index.max(), freq="H")
    missing = expected.difference(df.index)

    assert len(missing) == 0, f"Missing timestamps: {missing[:5]}"
```

```python
def mononotic_increasing (df):
    assert (df.index.is_monotonic_increasing)
    
def duplicated_index(df):
    assert df.index.duplicated().any() == False, "Duplicate timestamps found"
    
def missing_values(df):
    assert df.isna().sum() == 0, "Missing values found"
```

```python
def data_validation(df):
	negative_load(df)
	check_no_missing_hours(df)
	mononotic_increasing(df)
	duplicated_index(df)
	missing_values(df)
 
```
